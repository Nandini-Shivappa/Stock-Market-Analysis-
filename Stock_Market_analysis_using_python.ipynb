#  Moving Average Analysis

A moving average(MA) is a widely used indicator in technical analysis that helps smooth out the price action by filtering out the noise from random price fluctuation.

This method helps us to determine the trend.  

```python
# Moving average for different periods

ma_day = [10, 20, 50]

tech_stock_data = [AAPL, GOOG, MSFT, AMZN, NVDA]

for stock in tech_stock_data:
  for ma in ma_day:
    column_name = f"MA_for {ma}_day"
    stock[column_name] = stock['Close'].rolling(window=ma).mean()
```

```python
AAPL.head(50)
```

```python
AAPL[['Close', 'MA_for 10_day', 'MA_for 20_day', 'MA_for 50_day']].plot(legend=True, figsize=(14,6))

plt.show()
```

```python
# Plot MA of 20 days for different stocks

plt.figure(figsize=(20,8))

for i,stock in enumerate(tech_stock_data):
  stock['MA_for 20_day'].plot(label=tech_list[i]) # Corrected plt to plot

plt.legend()
plt.title('MA of 20 days for different stocks', fontsize=18)
plt.show()
```

```python
# Correlation between MSFT and GOOGLE stock

sns.jointplot(x='GOOG_Close', y='MSFT_Close', data=closing_df, kind='scatter', color='green')
plt.title("Correlation between the closing price of MSFT and GOOG")
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(closing_df)
```

```python
tech_returns = pd.DataFrame()
for stock, name in zip(tech_stock_data, tech_list):
    tech_returns[name] = stock['Daily Return']

sns.pairplot(tech_returns)
```

```python
returns_fig = sns.PairGrid(tech_returns.dropna())
returns_fig.map_upper(plt.scatter, color='purple')
returns_fig.map_lower(sns.kdeplot, cmap='cool_d')
returns_fig.map_diag(plt.hist,bins=30)
plt.show()
```

```python
price_fig = sns.PairGrid(closing_df.dropna())

price_fig.map_upper(plt.scatter, color='purple')
price_fig.map_lower(sns.kdeplot, cmap='viridis')
price_fig.map_diag(plt.hist, bins=30)

plt.show()
```

```python
corr = tech_returns.dropna().corr()
mask = np.zeros_like(corr)
# Corrected typo: changed triu_indicies_from to triu_indices_from
mask[np.triu_indices_from(mask)] = True
fig, ax = plt.subplots(figsize=(11,7))
cmap = sns.diverging_palette(220,10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, annot = True)
plt.title("Correlation between stocks")
plt.show()
```
